# Translations and Reviews in Iberoamerican Modernist Periodicals


 This dataset records translations and reviews of foreign works published in Spanish or Latin American modernist journals (understanding 'modernist' in the anglophone sense). Thus, it currently includes all translations or reviews from the following journals: La Revista (Barcelona) from 1915 to 1936; the second period of Proa (Buenos Aires) from 1924 to 1926, and Sur (Buenos Aires) from 1931 to 1939. 
 
http://openaccess.uoc.edu/webapps/o2/handle/10609/86485

## Traducciones

```python
from sankey import sankey
```

```python
languages = df["original_language_iso_1"].dropna().unique()
```

```python
languages = list(languages) + ['cat',]
```

```python
sankey(df[["original_language_iso_1", "dc_language_iso"]].dropna()["original_language_iso_1"],
       df[["original_language_iso_1", "dc_language_iso"]].dropna()["dc_language_iso"],
       colorDict=dict(zip(languages, sns.color_palette('Paired', n_colors=len(languages)))), )
```

## Red de editores

```python
from pyvis.network import Network
g = Network(notebook=True, height="750px", width="100%")
```

```python
sources = df[['dc_publisher', 'original_language_iso_1']].dropna()['dc_publisher']
targets = df[['dc_publisher', 'original_language_iso_1']].dropna()['original_language_iso_1']

edge_data = zip(sources, targets)

for e in edge_data:
    src = e[0]
    dst = e[1]
    w = 1

    g.add_node(src, src, title=src, color="red")
    g.add_node(dst, dst, title=dst, color="steelblue" )
    g.add_edge(src, dst, value=w, color='black')
```
